```python
import pandas as pd 
import numpy as np 

import plotly.express as px 
```

```python
raw_mse = pd.read_csv('training_curves/raw_training_mse_32.csv')
```

```python
pretrained_mse = pd.read_csv('training_curves/training_mse_32.csv')
```

```python
pretrained_mse['type'] = 'pretrained'
raw_mse['type']= 'not_pretrained' 
```

```python
import math
```

```python
def smooth(scalars: list[float], weight: float) -> list[float]:
    """
    EMA implementation according to
    https://github.com/tensorflow/tensorboard/blob/34877f15153e1a2087316b9952c931807a122aa7/tensorboard/components/vz_line_chart2/line-chart.ts#L699
    """
    last = 0
    smoothed = []
    num_acc = 0
    for next_val in scalars:
        last = last * weight + (1 - weight) * next_val
        num_acc += 1
        # de-bias
        debias_weight = 1
        if weight != 1:
            debias_weight = 1 - math.pow(weight, num_acc)
        smoothed_val = last / debias_weight
        smoothed.append(smoothed_val)

    return smoothed
```

```python
raw_mse['mse'] =  smooth(raw_mse.Value, weight=0.99)
pretrained_mse['mse'] =  smooth(pretrained_mse.Value, weight=0.99)
```

```python
df = pd.concat([raw_mse, pretrained_mse])
```
